## 3. rendering in Real Time

```python
#Connect to webcam
cap = cv2.VideoCapture(0)
# Loop through every frame until we close our webcam
while cap.isOpened(): 
    ret, frame = cap.read()
    
    # Show image 
    cv2.imshow('Webcam', frame)
    cv2.imwrite('webcamphoto_plant.jpg', frame)
    # Checks whether q has been hit and stops the loop
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

# Releases the webcam
cap.release()
# Closes the frame
cv2.destroyAllWindows()
```

```python
def take_photo_plant():
    #Connect to webcam
    cap = cv2.VideoCapture(0)
    # Loop through every frame until we close our webcam
    while cap.isOpened(): 
        ret, frame = cap.read()
        
        # Show image 
        cv2.imshow('Webcam', frame)
        cv2.imwrite('webcamphoto_plant.jpg', frame)
        # Checks whether q has been hit and stops the loop
        if cv2.waitKey(1) & 0xFF == ord('q'): 
            break

    # Releases the webcam
    cap.release()
    # Closes the frame
    cv2.destroyAllWindows()
```

```python
take_photo_plant()
```
